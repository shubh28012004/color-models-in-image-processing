RGB TO GRAYSCALE
1. Lightness method
    grayscale = (min(RGB) + max(RGB))/2
2. Average Method
    grayscale = (R+G+B)/2
3. lumniosity Method
    grayscale = 0.3*R + 0.59*G + 0.11*B

**Grayscale image to binary image **

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/Dataset/agri_0_1017.jpeg')
img
```

```python
#convert the input image to grayscale
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
gray
```

```python
#applying thresholding to convert grayscale to binary image
ret,thresh = cv2.threshold(gray,127,255,0)
ret,thresh
```

```python
#convert BGR to RGB so that it can be displayed using matplot
imgRGB = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
imgRGB
```

```python
#display original , grayscale  and binary images
plt.subplot(131),plt.imshow(imgRGB,cmap='gray'),plt.title('Original image '),plt.axis('off')
plt.subplot(132),plt.imshow(gray,cmap='gray'),plt.title('Gray image '),plt.axis('off')
plt.subplot(133),plt.imshow(thresh,cmap='gray'),plt.title('Binary image '),plt.axis('off')
plt.show()
```
